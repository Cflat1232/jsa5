# Read Assignment Document

This notebook extracts text from the attached Word document and displays the assignment requirements.

```python
import zipfile
from xml.etree import ElementTree as ET

path = r'C:\Users\Conner\Downloads\Assignment #5 (1).docx'
with zipfile.ZipFile(path) as docx:
    xml = docx.read('word/document.xml')
root = ET.fromstring(xml)
namespaces = {'w': 'http://schemas.openxmlformats.org/wordprocessingml/2006/main'}
texts = [node.text for node in root.findall('.//w:t', namespaces) if node.text]
print('
'.join(texts))
```
